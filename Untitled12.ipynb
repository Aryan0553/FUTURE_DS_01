```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
```

```python
#Rename Columns (Standard Format)
df.columns = df.columns.str.strip().str.replace(" ", "_").str.lower()
df.columns
```

```python
#Data Cleaning
df.isnull().sum()
```

```python
df.dropna(subset=["order_id", "order_date", "product_name", "quantity", "unit_price"], inplace=True)
```

```python
#fixed data type
df["order_date"] = pd.to_datetime(df["order_date"])
df["quantity"] = df["quantity"].astype(int)
df["unit_price"] = df["unit_price"].astype(float)
```

```python
#Remove Invalid Rows
df = df[(df["quantity"] > 0) & (df["unit_price"] > 0)]
```

```python
#Feature Engineering
#Create Sales & Profit
df["sales"] = df["quantity"] * df["unit_price"]
df["profit"] = df["sales"] * 0.20   # assumed 20% margin
```

```python
df["year"] = df["order_date"].dt.year
df["month"] = df["order_date"].dt.month
df["month_name"] = df["order_date"].dt.strftime("%b")
```

```python
#KPI Analysis
total_sales = df["sales"].sum()
total_profit = df["profit"].sum()
total_orders = df["order_id"].nunique()
avg_order_value = total_sales / total_orders

total_sales, total_profit, total_orders, avg_order_value
```

```python
#Monthly Sales Trend 
monthly_sales = (
    df.groupby(["year", "month_name"])["sales"]
    .sum()
    .reset_index()
)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=monthly_sales, x="month_name", y="sales", marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Best-Selling Products
top_products = (
    df.groupby("product_name")["sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
```

```python
plt.figure(figsize=(8,5))
top_products.sort_values().plot(kind="barh")
plt.title("Top 10 Best-Selling Products")
plt.xlabel("Total Sales")
plt.show()
```

```python
#High-Revenue Categories
category_sales = df.groupby("category")["sales"].sum().sort_values(ascending=False)
category_sales
```

```python
plt.figure(figsize=(6,6))
category_sales.plot(kind="pie", autopct="%1.1f%%", startangle=90)
plt.title("Revenue by Category")
plt.ylabel("")
plt.show()
```

```python
#Region-wise Sales
region_sales = df.groupby("region")["sales"].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(7,4))
region_sales.plot(kind="bar")
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Quantity vs Sales Relationship
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="quantity", y="sales")
plt.title("Quantity vs Sales")
plt.show()
```
